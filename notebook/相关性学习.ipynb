```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(50)
y = 2 * x + np.random.normal(0, 0.1, 50)
np.cov(x, y)[0, 1]
```

```python
Stock1 = get_price("000300.XSHG",start_date='2015-05-01',end_date='2015-08-01',fields = 'ClosingPx')
Stock2 = get_price("399006.XSHE",start_date='2015-05-01',end_date='2015-08-01',fields = 'ClosingPx')

plt.scatter(Stock1,Stock2)
plt.xlabel("601186.XSHG price")
plt.ylabel("601390.XSHG price")
plt.show()
print("the corrlation for two stocks is: ")
Stock2.corr(Stock1)
```
